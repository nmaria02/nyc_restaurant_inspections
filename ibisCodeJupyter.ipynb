```python
sub=t.select("CAMIS",    
	"DBA", 
	"CUISINE DESCRIPTION", 
	"INSPECTION DATE",
	"VIOLATION CODE",
	"VIOLATION DESCRIPTION",
	"CRITICAL FLAG")
```

```python
sub.select("CRITICAL FLAG").distinct()
```

```python
sub.filter(sub["CRITICAL FLAG"]=="Critical")
```

```python
pizza=sub.filter((sub["CRITICAL FLAG"]=="Critical") & 
                     (sub["CUISINE DESCRIPTION"]=="Pizza"))

```

```python
pizza_sort=pizza.order_by("DBA")
```
